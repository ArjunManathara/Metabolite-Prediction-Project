```python
water_regions1 = [
    (1440, 1460),  # Around 1450 nm
    (1900, 2070),  # Around 1940 nm
]
def filterRange(col,water_regions):
    for i in range(len(water_regions)):
        if(col>=water_regions[i][0] and col<=water_regions[i][1]):
            return 1
    return 0

vals = [filterRange(i,water_regions1) for i in wavelengths]
plt.plot(wavelengths,vals)
#plt.show()

plt.plot(wavelengths,water_spectra.iloc[0])
plt.title('Water spectra peak approximation')
plt.show()
```

```python
def waterPreprocess(df,water_spectra):
    ''' This function generates an array of indices based on the given range of water peaks and for each sample, it
    calculates the average absorbance value in these regions and compares it with the average absorbance value of water in
    these regions. The minimum ratio is considered and the sample is subtracted with the water spectra scaled to this ratio.
    '''
    water_regions = [
        (1440, 1460),  # Around 1450 nm
        (1900, 2070),  # Around 1940 nm
    ]
    water_indices = [np.where((np.array(wavelengths).astype(float) >= region[0]) \
                              & (np.array(wavelengths).astype(float) <= region[1]))[0].astype(int) \
                     for region in water_regions]
    resultant_spectra = np.zeros_like(df)
    for sample_idx in range(df.shape[0]):
        scaling_ratio = 2
        for indiceRange in water_indices:
            sample_sum=0
            water_sum=0
            for indexVal in indiceRange:
                sample_sum += df.iloc[sample_idx,indexVal]
                water_sum += water_spectra.iloc[0,indexVal]
            scaling_ratio = min(scaling_ratio,sample_sum/water_sum)
        resultant_spectra[sample_idx] = np.array(df.iloc[sample_idx]) - scaling_ratio*np.array(water_spectra.iloc[0])
    return resultant_spectra
```

```python
def multiplicative_scatter_correction(X, reference=None):
    """
    Apply Multiplicative Scatter Correction (MSC) to each row in the dataset X.
    X: The input data, where each row is a spectrum.
    reference: The reference spectrum (default is the mean of X).
    
    Returns:
        MSC corrected data.
    """
    # If no reference is provided, use the mean of X as the reference
    if reference is None:
        reference = np.mean(X, axis=0)
    
    # Center the data by subtracting the mean of each spectrum
    centered_data = X - np.mean(X, axis=1, keepdims=True)
    
    # Compute the scaling factor based on the reference spectrum
    scaling_factor = np.dot(centered_data, reference) / np.dot(reference, reference)
    
    # Apply MSC correction
    corrected_data = centered_data - scaling_factor[:, np.newaxis] * reference
    
    return corrected_data
```

```python
# Savitzky Golay filtering
from scipy.signal import savgol_filter

# Apply Savitzky-Golay filtering (1st derivative)
def savGol(df):
    window_size = 5  # Window size (choose an odd number)
    poly_order = 3    # Polynomial order
    return savgol_filter(df, window_length=window_size, polyorder=poly_order, deriv=1)
```

### Outlier detection and removal

```python
import numpy as np
import scipy.stats as stats
from sklearn.decomposition import PCA

def hotelling_t2(X_pca, confidence=0.95):
    """Computes Hotelling's T² for PCA transformed data."""
    mean_vector = np.mean(X_pca, axis=0)  # Mean of PCA components
    cov_matrix = np.cov(X_pca.T)  # Covariance matrix of PCA space
    inv_cov_matrix = np.linalg.inv(cov_matrix)  # Inverse covariance matrix

    t2_values = []
    for x in X_pca:
        diff = x - mean_vector
        t2 = diff.T @ inv_cov_matrix @ diff  # Hotelling's T² formula
        t2_values.append(t2)

    # Set threshold based on F-distribution
    n, p = X_pca.shape  # n = samples, p = PCA components
    threshold = (p * (n - 1) / (n - p)) * stats.f.ppf(confidence, dfn=p, dfd=n-p)

    outliers = np.where(np.array(t2_values) > threshold)[0]  # Get indices of outliers

    return np.array(t2_values), threshold, outliers

# Apply PCA
pca = PCA(n_components=3)
ld_pca = pca.fit_transform(ld)  # X = Your preprocessed spectral data

# Compute Hotelling's T²
t2_values, t2_threshold, outliers_t2 = hotelling_t2(ld_pca, confidence=0.95)

print(f"Number of outliers detected by Hotelling’s T²: {len(outliers_t2)}")
print(f"Outlier indices: {outliers_t2}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.base import BaseEstimator, TransformerMixin

class WaterPreprocessing(BaseEstimator, TransformerMixin):
    def __init__(self,water_spectra):
        self.water_spectra = water_spectra
        self.water_spectra = water_spectra
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        # Apply water preprocessing to X
        return waterPreprocess(X,self.water_spectra)

class SavitzkyGolayFiltering(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        # Apply Savitzky-Golay filtering
        return savGol(X)
    
class NoPreprocessing(BaseEstimator, TransformerMixin):
    def fit(self,X, y=None):
        return self
    def transform(self, X):
        return np.array(X)
    
class MSC(BaseEstimator, TransformerMixin):
    def fit(self,X, y = None):
        return self
    def transform(self,X):
        return multiplicative_scatter_correction(X)
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

def create_pca_pipeline(preprocessing_pipeline, use_unlabelled_data=False):
    # PCA Step
    scaler_pipeline = Pipeline(steps=[
        ('scaler', StandardScaler()),  # Fit scaler on X_train, transform X_test
    ])
    
    pca_pipeline = Pipeline(steps=[
        ('pca', PCA(n_components=20))  # Apply PCA
    ])
    
    # Full pipeline with optional unlabelled data handling
    def full_pipeline(X_train, X_test, unlabelled_data=None):
        # Apply preprocessing steps to X_train and X_test
        X_train_transformed = preprocessing_pipeline.fit_transform(X_train)
        X_test_transformed = preprocessing_pipeline.transform(X_test)
        
        # Apply StandardScaler to X_train
        X_train_scaled = scaler_pipeline.fit_transform(X_train_transformed)
        
        # Apply the same scaler to X_test
        X_test_scaled = scaler_pipeline.transform(X_test_transformed)
        
        # Handle unlabelled data if provided
        if use_unlabelled_data and unlabelled_data is not None:
            unlabelled_data_transformed = preprocessing_pipeline.transform(unlabelled_data)
            unlabelled_data_scaled = scaler_pipeline.transform(unlabelled_data_transformed)
            
            # Concatenate X_train_scaled with unlabelled data for PCA fitting
            combined_data_scaled = np.vstack((X_train_scaled, unlabelled_data_scaled))
        else:
            # If no unlabelled data, only use X_train_scaled for PCA fitting
            combined_data_scaled = X_train_scaled
        
        # Apply PCA on the combined data (X_train_scaled + unlabelled data, if any)
        pca_pipeline.fit(combined_data_scaled)  # Fit PCA on the scaled data
        X_train_pca = pca_pipeline.transform(X_train_scaled)  # Transform only X_train_scaled
        X_test_pca = pca_pipeline.transform(X_test_scaled)    # Transform only X_test_scaled
        
        return X_train_pca, X_test_pca, pca_pipeline

    return full_pipeline
```

```python
def plot_pca_cumulative_variance(pca_pipeline):
    # Cumulative variance explained by each component
    cumulative_variance = np.cumsum(pca_pipeline.named_steps['pca'].explained_variance_ratio_)
    
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
    plt.title('Cumulative Explained Variance by PCA Components')
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.axhline(0.98,color='b',linestyle='--',label='98% Threshold')
    plt.axhline(0.95, color="r", linestyle="--", label="95% Threshold")
    plt.grid(True)
    plt.show()
```

### Results for Linear models : Linear regression, Ridge Regression, Lasso Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import make_scorer, r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Store the results for model comparison
model_comparison_results = []

# Define the models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso()
}

# Loop through each model, train it, and evaluate it
for model_name, model in models.items():
    print(f"Training and evaluating {model_name}...")

    # Store the R² scores for cross-validation, training, and testing errors
    model_results = []

    # Loop through each metabolite (column of y)
    for metabolite in y_train.columns:
        y_metabolite_train = y_train[metabolite]
        y_metabolite_test = y_test[metabolite]

        # Create and fit the model pipeline (with scaling)
        model_pipeline = Pipeline(steps=[
            ('model', model)
        ])

        # Perform cross-validation to get R² scores
        r2_scorer = make_scorer(r2_score)
        cv_scores = cross_val_score(model_pipeline, X_train_pca[:,:pca_components], y_metabolite_train, cv=5, scoring=r2_scorer)

        # Calculate the average R² score from cross-validation
        avg_r2_cv = cv_scores.mean()

        # Fit the model on training data and calculate training R² score
        model_pipeline.fit(X_train_pca[:,:pca_components], y_metabolite_train)
        y_train_pred = model_pipeline.predict(X_train_pca[:,:pca_components])
        train_r2 = r2_score(y_metabolite_train, y_train_pred)

        # Calculate the testing R² score
        y_test_pred = model_pipeline.predict(X_test_pca[:,:pca_components])
        test_r2 = r2_score(y_metabolite_test, y_test_pred)

        # Store the results: Model, Metabolite, Training R², Testing R², CV R²
        model_results.append([model_name, metabolite, train_r2, test_r2, avg_r2_cv])

    # Append the results for this model
    model_comparison_results.extend(model_results)

# Convert the results into a DataFrame for easier viewing
columns = ['Model', 'Metabolite', 'Train R²', 'Test R²', 'CV R²']
model_comparison_df = pd.DataFrame(model_comparison_results, columns=columns)

# Display the results in a table-like plot (heatmap)
plt.figure(figsize=(10, 6))
# Pivot the data for better visualization
pivot_df = model_comparison_df.pivot(index='Metabolite', columns='Model', values=['Train R²', 'Test R²', 'CV R²'])

# Plot the heatmap using seaborn
sns.heatmap(pivot_df, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Model Comparison on Training, Testing, and Cross-validation R² Scores")
plt.tight_layout()
plt.show()
```
